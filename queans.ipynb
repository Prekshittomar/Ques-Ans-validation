```python
import json

# Open the JSON file and load it into a Python object (e.g., dictionary)
with open('D:\\qa_pairs_V2.json', 'r') as file:
    data = json.load(file)
```

```python
image_index=[]
qid=[]
question_string=[]
template=[]
type=[]
answer=[]
question_id=[]

i=0
```

```python
for i in range(4360648):
    image_index.append(data['qa_pairs'][i]['image_index'])
    qid.append(data['qa_pairs'][i]['qid'])
    question_string.append(data['qa_pairs'][i]['question_string'])
    template.append(data['qa_pairs'][i]['template'])
    type.append(data['qa_pairs'][i]['type'])
    answer.append(data['qa_pairs'][i]['answer'])
    question_id.append(data['qa_pairs'][i]['question_id'])
```

```python
df_2 = pd.DataFrame({
    'image_index':image_index,
    'qid':qid,
    'question_string':question_string,
    'template':template,
    'type':type,
    'answer':answer,
    'question_id':question_id
})
```

```python
new_df = pd.merge(df_1,df_2,on='image_index')
```
